Find the potato variety with the shortest days to maturity, then list the variety, where to buy seed potatoes, how to prepare, as well as a photo. 

```python
from PIL import Image
min_harvest = df['days_to_maturity'].min()
print("Days to Maturity: " + str(min_harvest))
min_list = df['days_to_maturity'].tolist()
variety_list = df['variety'].tolist()
seller_list = df['seller'].tolist()
file_list = df['file'].tolist()
type_list = df['type'].tolist()
#print(min_list)
#print(len(min_list))
#print(variety_list)
#print(len(variety_list))
for i in range(len(min_list)):
  if i == 0:
    min_val=min_list[i]
    min_loc=i
  elif min_list[i] < min_val:
    min_val=min_list[i]
    min_loc=i
#print(min_val)
#print(min_loc)
#print(min_list[min_loc])
print("Variety: " + str(variety_list[min_loc]))
print("Seller(s): " + str(seller_list[min_loc]))

if str(type_list[min_loc]) == "all purpose": 
  print("Recommended preparation: STICK 'EM IN A STEW!")
elif str(type_list[min_loc]) == "waxy": 
  print("Recommended preparation: MASH 'EM!")
elif str(type_list[min_loc]) == "starchy": 
  print("Recommended preparation: BOIL 'EM!")

img = Image.open(str(file_list[min_loc]))
img.show()

```
